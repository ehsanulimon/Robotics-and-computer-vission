```python
from PIL import Image

# Open an image
image_path = 'D:/4.3/new sir/images.jpg'
image = Image.open(image_path)

# Separate channels
red_channel, green_channel, blue_channel = image.split()

# Extract pixel values
pixels_red = list(red_channel.getdata())
pixels_green = list(green_channel.getdata())
pixels_blue = list(blue_channel.getdata())

# Display or use the pixel values as needed
print("Red Channel Pixels:", pixels_red[:10])  # Displaying first 10 pixels as an example
print("Green Channel Pixels:", pixels_green[:10])
print("Blue Channel Pixels:", pixels_blue[:10])
```
